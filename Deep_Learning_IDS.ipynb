```python
"""
Deep Learning IDS - Model Training Script for Google Colab
Train CNN and LSTM models with extreme class imbalance handling
Save trained models and preprocessing artifacts for deployment
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import BorderlineSMOTE
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.utils import to_categorical
import pickle
import warnings
warnings.filterwarnings('ignore')

print("TensorFlow version:", tf.__version__)
print("GPU Available:", len(tf.config.list_physical_devices('GPU')) > 0)

# ============================================================================
# DATA LOADING AND PREPROCESSING
# ============================================================================

def load_nsl_kdd_data(train_path, test_path):
    """Load NSL-KDD dataset"""
    columns = ['duration', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes',
               'land', 'wrong_fragment', 'urgent', 'hot', 'num_failed_logins', 'logged_in',
               'num_compromised', 'root_shell', 'su_attempted', 'num_root', 'num_file_creations',
               'num_shells', 'num_access_files', 'num_outbound_cmds', 'is_host_login',
               'is_guest_login', 'count', 'srv_count', 'serror_rate', 'srv_serror_rate',
               'rerror_rate', 'srv_rerror_rate', 'same_srv_rate', 'diff_srv_rate',
               'srv_diff_host_rate', 'dst_host_count', 'dst_host_srv_count',
               'dst_host_same_srv_rate', 'dst_host_diff_srv_rate', 'dst_host_same_src_port_rate',
               'dst_host_srv_diff_host_rate', 'dst_host_serror_rate', 'dst_host_srv_serror_rate',
               'dst_host_rerror_rate', 'dst_host_srv_rerror_rate', 'attack_type', 'difficulty']

    train_df = pd.read_csv(train_path, names=columns)
    test_df = pd.read_csv(test_path, names=columns)

    print(f"Training samples: {len(train_df)}")
    print(f"Test samples: {len(test_df)}")

    return train_df, test_df

def map_attack_categories(df):
    """Map specific attacks to 5 main categories"""
    attack_mapping = {
        'normal': 'Normal',
        'neptune': 'DoS', 'smurf': 'DoS', 'pod': 'DoS', 'teardrop': 'DoS',
        'land': 'DoS', 'back': 'DoS', 'apache2': 'DoS', 'udpstorm': 'DoS',
        'processtable': 'DoS', 'mailbomb': 'DoS',
        'portsweep': 'Probe', 'ipsweep': 'Probe', 'nmap': 'Probe', 'satan': 'Probe',
        'saint': 'Probe', 'mscan': 'Probe',
        'guess_passwd': 'R2L', 'ftp_write': 'R2L', 'imap': 'R2L', 'phf': 'R2L',
        'multihop': 'R2L', 'warezmaster': 'R2L', 'warezclient': 'R2L', 'spy': 'R2L',
        'xlock': 'R2L', 'xsnoop': 'R2L', 'snmpguess': 'R2L', 'snmpgetattack': 'R2L',
        'httptunnel': 'R2L', 'sendmail': 'R2L', 'named': 'R2L',
        'buffer_overflow': 'U2R', 'loadmodule': 'U2R', 'rootkit': 'U2R', 'perl': 'U2R',
        'sqlattack': 'U2R', 'xterm': 'U2R', 'ps': 'U2R'
    }

    df['category'] = df['attack_type'].map(attack_mapping)
    df['category'] = df['category'].fillna('Normal')
    return df

def engineer_features(df):
    """Create domain-specific features"""
    df['failed_login_ratio'] = df['num_failed_logins'] / (df['count'] + 1)
    df['guest_anomaly'] = df['is_guest_login'] * df['num_file_creations']
    df['service_error_pattern'] = df['srv_serror_rate'] * df['dst_host_srv_serror_rate']
    df['data_transfer_anomaly'] = df['src_bytes'] / (df['dst_bytes'] + 1)
    df['privilege_escalation_score'] = (df['su_attempted'] * 3 +
                                        df['root_shell'] * 5 +
                                        df['num_root'] * 2)
    df['root_access_pattern'] = df['root_shell'] * df['num_compromised']
    df['file_system_anomaly'] = (df['num_file_creations'] +
                                  df['num_shells'] +
                                  df['num_access_files'])
    return df

def preprocess_data(train_df, test_df):
    """Complete preprocessing pipeline"""
    train_df = map_attack_categories(train_df)
    test_df = map_attack_categories(test_df)

    train_df = engineer_features(train_df)
    test_df = engineer_features(test_df)

    drop_cols = ['attack_type', 'difficulty']
    train_df = train_df.drop(columns=drop_cols)
    test_df = test_df.drop(columns=drop_cols)

    # Store categorical columns for later use
    categorical_cols = ['protocol_type', 'service', 'flag']
    label_encoders = {}

    for col in categorical_cols:
        le = LabelEncoder()
        train_df[col] = le.fit_transform(train_df[col].astype(str))
        test_df[col] = le.transform(test_df[col].astype(str))
        label_encoders[col] = le

    X_train = train_df.drop('category', axis=1).values
    y_train = train_df['category'].values
    X_test = test_df.drop('category', axis=1).values
    y_test = test_df['category'].values

    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    le = LabelEncoder()
    y_train_encoded = le.fit_transform(y_train)
    y_test_encoded = le.transform(y_test)

    feature_names = train_df.drop('category', axis=1).columns.tolist()

    return X_train, X_test, y_train_encoded, y_test_encoded, le, scaler, label_encoders, feature_names

# ============================================================================
# EXTREME SMOTE
# ============================================================================

def apply_extreme_smote(X_train, y_train):
    """Apply extreme oversampling"""
    unique, counts = np.unique(y_train, return_counts=True)
    class_distribution = dict(zip(unique, counts))

    print("\n--- Original Class Distribution ---")
    for cls, cnt in class_distribution.items():
        print(f"Class {cls}: {cnt} samples ({cnt/len(y_train)*100:.2f}%)")

    max_samples = max(counts)
    target_samples = int(max_samples * 0.8)

    sampling_strategy = {cls: max(cnt, target_samples)
                        for cls, cnt in class_distribution.items()}

    min_samples = min(counts)
    k_neighbors = 1 if min_samples < 6 else 5

    print(f"\nApplying SMOTE with k_neighbors={k_neighbors}")
    smote = BorderlineSMOTE(sampling_strategy=sampling_strategy,
                           k_neighbors=k_neighbors,
                           random_state=42)

    X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

    print("\n--- After SMOTE Class Distribution ---")
    unique, counts = np.unique(y_resampled, return_counts=True)
    for cls, cnt in zip(unique, counts):
        print(f"Class {cls}: {cnt} samples ({cnt/len(y_resampled)*100:.2f}%)")

    return X_resampled, y_resampled

# ============================================================================
# FOCAL LOSS
# ============================================================================

def focal_loss(gamma=2.0, alpha=0.25):
    """Focal loss implementation"""
    def focal_loss_fixed(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1.0 - epsilon)
        cross_entropy = -y_true * tf.math.log(y_pred)
        loss = alpha * tf.pow(1 - y_pred, gamma) * cross_entropy
        return tf.reduce_mean(tf.reduce_sum(loss, axis=-1))
    return focal_loss_fixed

# ============================================================================
# MODEL ARCHITECTURES
# ============================================================================

def build_enhanced_cnn(input_shape, num_classes):
    """Build enhanced CNN model"""
    inputs = layers.Input(shape=input_shape)
    x = layers.Reshape((input_shape[0], 1))(inputs)

    # Multi-scale convolutions
    conv3 = layers.Conv1D(64, 3, padding='same', activation='relu')(x)
    conv3 = layers.BatchNormalization()(conv3)

    conv5 = layers.Conv1D(64, 5, padding='same', activation='relu')(x)
    conv5 = layers.BatchNormalization()(conv5)

    conv7 = layers.Conv1D(64, 7, padding='same', activation='relu')(x)
    conv7 = layers.BatchNormalization()(conv7)

    x = layers.Concatenate()([conv3, conv5, conv7])
    x = layers.MaxPooling1D(2)(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Conv1D(128, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2)(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Conv1D(256, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dropout(0.4)(x)

    x = layers.Dense(256, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)

    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.4)(x)

    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = models.Model(inputs=inputs, outputs=outputs, name='Enhanced_CNN')
    return model

def build_enhanced_lstm(input_shape, num_classes):
    """Build enhanced LSTM model"""
    inputs = layers.Input(shape=input_shape)
    x = layers.Reshape((input_shape[0], 1))(inputs)

    x = layers.Bidirectional(layers.LSTM(128, return_sequences=True))(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Bidirectional(layers.LSTM(64, return_sequences=True))(x)
    x = layers.Dropout(0.3)(x)

    # Attention
    attention = layers.Dense(1, activation='tanh')(x)
    attention = layers.Flatten()(attention)
    attention = layers.Activation('softmax')(attention)
    attention = layers.RepeatVector(128)(attention)
    attention = layers.Permute([2, 1])(attention)

    x = layers.Multiply()([x, attention])
    x = layers.GlobalAveragePooling1D()(x)

    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.4)(x)

    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dropout(0.3)(x)

    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = models.Model(inputs=inputs, outputs=outputs, name='Enhanced_LSTM')
    return model

# ============================================================================
# TRAINING
# ============================================================================

def compute_extreme_class_weights(y_train, boost_factor=10):
    """Compute extreme class weights"""
    classes = np.unique(y_train)
    base_weights = compute_class_weight('balanced', classes=classes, y=y_train)

    unique, counts = np.unique(y_train, return_counts=True)
    class_counts = dict(zip(unique, counts))

    min_count = min(counts)
    extreme_weights = {}

    for i, cls in enumerate(classes):
        if class_counts[cls] < min_count * 5:
            extreme_weights[cls] = base_weights[i] * boost_factor
        else:
            extreme_weights[cls] = base_weights[i]

    print("\n--- Extreme Class Weights ---")
    for cls, weight in extreme_weights.items():
        print(f"Class {cls}: {weight:.2f}")

    return extreme_weights

def train_model(model, X_train, y_train, X_val, y_val, class_weights,
                model_name='model', epochs=50):
    """Train model"""
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss=focal_loss(gamma=2.0, alpha=0.25),
        metrics=['accuracy']
    )

    early_stop = callbacks.EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True
    )

    reduce_lr = callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-6
    )

    history = model.fit(
        X_train, y_train,
        batch_size=64,
        epochs=epochs,
        validation_data=(X_val, y_val),
        class_weight=class_weights,
        callbacks=[early_stop, reduce_lr],
        verbose=1
    )

    return history

# ============================================================================
# THRESHOLD OPTIMIZATION
# ============================================================================

def optimize_thresholds(model, X_val, y_val, label_encoder):
    """Optimize thresholds for each class"""
    y_pred_proba = model.predict(X_val)
    best_thresholds = {}
    class_names = label_encoder.classes_

    for class_idx, class_name in enumerate(class_names):
        best_f1 = 0
        best_threshold = 0.5

        if class_name in ['R2L', 'U2R']:
            threshold_range = np.arange(0.01, 0.2, 0.01)
        else:
            threshold_range = np.arange(0.3, 0.7, 0.05)

        for threshold in threshold_range:
            y_pred_class = (y_pred_proba[:, class_idx] > threshold).astype(int)
            y_true_binary = (y_val == class_idx).astype(int)
            f1 = f1_score(y_true_binary, y_pred_class, zero_division=0)

            if f1 > best_f1:
                best_f1 = f1
                best_threshold = threshold

        best_thresholds[class_name] = best_threshold
        print(f"{class_name}: Threshold={best_threshold:.3f}, F1={best_f1:.3f}")

    return best_thresholds

# ============================================================================
# EVALUATION
# ============================================================================

def evaluate_model(model, X_test, y_test, label_encoder, model_name='Model'):
    """Evaluate model on test set"""
    y_pred = np.argmax(model.predict(X_test), axis=1)

    print(f"\n{'='*60}")
    print(f"{model_name} - Classification Report")
    print(f"{'='*60}")
    print(classification_report(y_test, y_pred,
                                target_names=label_encoder.classes_,
                                digits=4))

    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=label_encoder.classes_,
                yticklabels=label_encoder.classes_)
    plt.title(f'{model_name} - Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig(f'{model_name.lower().replace(" ", "_")}_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

    return y_pred

def plot_training_history(history, model_name='Model'):
    """Plot training history"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    ax1.plot(history.history['loss'], label='Train Loss', linewidth=2)
    ax1.plot(history.history['val_loss'], label='Val Loss', linewidth=2)
    ax1.set_title(f'{model_name} - Loss', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    ax2.plot(history.history['accuracy'], label='Train Accuracy', linewidth=2)
    ax2.plot(history.history['val_accuracy'], label='Val Accuracy', linewidth=2)
    ax2.set_title(f'{model_name} - Accuracy', fontsize=14, fontweight='bold')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(f'{model_name.lower().replace(" ", "_")}_training_history.png', dpi=300, bbox_inches='tight')
    plt.show()

# ============================================================================
# MAIN TRAINING PIPELINE
# ============================================================================

def main():
    """Main training pipeline"""
    print("="*70)
    print("Deep Learning IDS - Training Pipeline")
    print("="*70)

    # 1. Load data
    print("\n[1/10] Loading NSL-KDD dataset...")
    train_df, test_df = load_nsl_kdd_data('/content/KDDTrain+.txt', '/content/KDDTest+.txt')

    # 2. Preprocess
    print("\n[2/10] Preprocessing data...")
    X_train, X_test, y_train, y_test, label_encoder, scaler, label_encoders, feature_names = preprocess_data(train_df, test_df)

    # 3. Split train/val
    print("\n[3/10] Splitting train/validation...")
    X_train, X_val, y_train, y_val = train_test_split(
        X_train, y_train, test_size=0.15, stratify=y_train, random_state=42
    )

    # 4. Apply SMOTE
    print("\n[4/10] Applying extreme SMOTE...")
    X_train_balanced, y_train_balanced = apply_extreme_smote(X_train, y_train)

    # 5. Compute class weights
    print("\n[5/10] Computing extreme class weights...")
    class_weights = compute_extreme_class_weights(y_train_balanced, boost_factor=10)

    # Convert to categorical
    num_classes = len(label_encoder.classes_)
    y_train_cat = to_categorical(y_train_balanced, num_classes)
    y_val_cat = to_categorical(y_val, num_classes)
    y_test_cat = to_categorical(y_test, num_classes)

    # 6. Train CNN
    print("\n[6/10] Training Enhanced CNN...")
    cnn_model = build_enhanced_cnn((X_train_balanced.shape[1],), num_classes)
    print(f"CNN Parameters: {cnn_model.count_params():,}")

    cnn_history = train_model(
        cnn_model, X_train_balanced, y_train_cat,
        X_val, y_val_cat, class_weights,
        model_name='CNN', epochs=50
    )

    plot_training_history(cnn_history, 'Enhanced CNN')

    # 7. Train LSTM
    print("\n[7/10] Training Enhanced LSTM...")
    lstm_model = build_enhanced_lstm((X_train_balanced.shape[1],), num_classes)
    print(f"LSTM Parameters: {lstm_model.count_params():,}")

    lstm_history = train_model(
        lstm_model, X_train_balanced, y_train_cat,
        X_val, y_val_cat, class_weights,
        model_name='LSTM', epochs=50
    )

    plot_training_history(lstm_history, 'Enhanced LSTM')

    # 8. Optimize thresholds
    print("\n[8/10] Optimizing classification thresholds...")
    print("\n--- CNN Thresholds ---")
    cnn_thresholds = optimize_thresholds(cnn_model, X_val, y_val, label_encoder)

    print("\n--- LSTM Thresholds ---")
    lstm_thresholds = optimize_thresholds(lstm_model, X_val, y_val, label_encoder)

    # 9. Evaluate
    print("\n[9/10] Evaluating models on test set...")
    evaluate_model(cnn_model, X_test, y_test, label_encoder, 'Enhanced CNN')
    evaluate_model(lstm_model, X_test, y_test, label_encoder, 'Enhanced LSTM')

    # Ensemble
    print("\n--- Ensemble Predictions ---")
    cnn_proba = cnn_model.predict(X_test)
    lstm_proba = lstm_model.predict(X_test)
    ensemble_proba = (cnn_proba + lstm_proba) / 2
    ensemble_pred = np.argmax(ensemble_proba, axis=1)

    print(classification_report(y_test, ensemble_pred,
                                target_names=label_encoder.classes_,
                                digits=4))

    # 10. Save everything
    print("\n[10/10] Saving models and artifacts...")

    # Save models
    cnn_model.save('cnn_ids_model.h5')
    lstm_model.save('lstm_ids_model.h5')

    # Save preprocessing artifacts
    artifacts = {
        'scaler': scaler,
        'label_encoder': label_encoder,
        'label_encoders': label_encoders,
        'feature_names': feature_names,
        'cnn_thresholds': cnn_thresholds,
        'lstm_thresholds': lstm_thresholds,
        'num_features': X_train.shape[1]
    }

    with open('preprocessing_artifacts.pkl', 'wb') as f:
        pickle.dump(artifacts, f)

    print("\n" + "="*70)
    print("âœ… Training Complete!")
    print("="*70)
    print("\nSaved files:")
    print("  â€¢ cnn_ids_model.h5")
    print("  â€¢ lstm_ids_model.h5")
    print("  â€¢ preprocessing_artifacts.pkl")
    print("\nDownload these files to use with the Flask application.")
    print("="*70)

    try:
        from google.colab import files
        print("\nğŸ“¥ Downloading files to your local system...")
        files.download('cnn_ids_model.h5')
        files.download('lstm_ids_model.h5')
        files.download('preprocessing_artifacts.pkl')
    except Exception as e:
        print("\nâš ï¸ Download skipped (not running in Colab). You can manually download the files from the working directory.")
        print(f"Error: {e}")

    print("="*70)

if __name__ == "__main__":
    main()
```
